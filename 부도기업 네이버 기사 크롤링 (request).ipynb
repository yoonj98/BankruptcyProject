## 부도기업 네이버뉴스 크롤링
#### By BeautifulSoup

```python
import re
import pandas as pd
import numpy as np
import requests
from bs4 import BeautifulSoup
```

```python
data=pd.read_csv('C:\\Users\\yoonj\\Desktop\\크롤링\\collapse\\대체기업 빈칸제외.csv', encoding='CP949')
keywords=data['대체기업명']
datamart=[]

start=list(np.array(data['부실6개월전'].tolist()))
end=list(np.array(data['부실발생일'].tolist()))

for i in range(0,len(start)):
    start[i]=re.sub("-", ".", start[i])

for i in range(0,len(end)):
    end[i]=re.sub("-", ".", end[i])
```

```python
data
```

```python
page=1
maxpage_t =(50-1)*10+1  #최대 50 페이지까지 

for i in range(0, len(data)):   #부도 기업명
    s_from = start[i].replace(".","")
    e_to = end[i].replace(".","")
    
    while page<maxpage_t:
        #키워드 별 페이지
        url = "https://search.naver.com/search.naver?where=news&query=" + keywords[i] + "&sort=0&ds=" + start[i] + "&de=" + end[i] + "&nso=so%3Ar%2Cp%3Afrom" + s_from + "to" + e_to + "%2Ca%3A&start=" + str(page)

        req = requests.get(url, headers={'User-Agent':'Mozilla/5.0'})
        cont = req.content
        soup = BeautifulSoup(cont, 'html.parser')
        
        for urls in soup.select(".info"):  #네이버 뉴스로 접근
            try :
                if urls["href"].startswith("https://news.naver.com"):  #네이버 뉴스인 경우
                    url1=urls["href"]

                    empty=[]

                    req1 = requests.get(url1, headers={'User-Agent':'Mozilla/5.0'})
                    html=req1.text
                    soup1 = BeautifulSoup(html, 'html.parser')

                    empty.append(keywords[i])
                    title=soup1.select('h3#articleTitle')[0].text                                    
                    empty.append(title)

                    text = soup1.select('#articleBodyContents')[0].get_text().replace('\n', " ")
                    btext = text.replace("// flash 오류를 우회하기 위한 함수 추가 function _flash_removeCallback() {}", "")                
                    empty.append(btext)
                    datamart.append(empty)   #병합
                    print("저장 완료")

            except Exception as e:
                print(e)
                continue
                
        page += 10  #페이지 넘김
        
    page=1  #페이지 리셋

df=pd.DataFrame(datamart)
```
