```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.select import Select
from selenium.webdriver.chrome.options import Options
from time import sleep
import re
import pandas as pd
import numpy as np
```

```python
data=pd.read_csv('C:\\Users\\yoonj\\Desktop\\크롤링\\collapse\\부도전처리.csv', encoding='CP949')
keywords=data['기업명']
datamart=[]

start=list(np.array(data['부실6개월전'].tolist()))
end=list(np.array(data['부실발생일'].tolist()))
quotes='"'

for i in range(0,len(start)):
    start[i]=re.sub("-", ".", start[i])

for i in range(0,len(end)):
    end[i]=re.sub("-", ".", end[i])
```

```python
url = ""
flag=True

driver = webdriver.Chrome()  #드라이버 킴
driver.get('https://news.naver.com/')

for i in range(0, len(data)): #각 부도기업 검색
    driver.find_element_by_name('query').send_keys(quotes)   #검색용 키워드 입력
    driver.find_element_by_name('query').send_keys(keywords[i])   #검색용 키워드 입력
    driver.find_element_by_name('query').send_keys(quotes)   #검색용 키워드 입력
    driver.find_element_by_name('query').submit()   #검색용 키워드 검색
    driver.switch_to.window(driver.window_handles[1])   #열린 창으로 페이지 이동
    
    if(flag == True):
        flag=False
        driver.find_element_by_link_text('검색옵션').send_keys(Keys.ENTER)  #검색 옵션 버튼 클릭
    
    driver.find_element_by_xpath('/html/body/div[3]/div[1]/div[2]/div/ul/li[2]/a').send_keys(Keys.ENTER)   #기간 검색 클릭
    driver.find_element_by_xpath('/html/body/div[3]/div[1]/div[2]/div/ul/li[2]/div/div[2]/div[1]/div[1]/input').send_keys(start[i])   #시작기간 클릭
    driver.find_element_by_xpath('/html/body/div[3]/div[1]/div[2]/div/ul/li[2]/div/div[2]/div[1]/div[2]/input').send_keys(end[i])   #끝기간 클릭
    driver.find_element_by_xpath('/html/body/div[3]/div[1]/div[2]/div/ul/li[2]/div/div[2]/span/button').send_keys(Keys.ENTER)   #적용 버튼    
    sleep(1)
    
    while True:  #기간 내 마지막 페이지까지 실행
        for j in range(1, 11):  #한 페이지 내 뉴스 접근
            try:
                page='/html/body/div[3]/div[2]/div/div[1]/section[2]/div/div[3]/ul/li['+str(j)+']/div[1]/div/div[1]/div/a[2]'  #이동 1
                page1='/html/body/div[3]/div[2]/div/div[1]/section[2]/div/div[3]/ul/li['+str(j)+']/div/div/div[1]/div/a[2]'   #이동 2       
                try:
                    url=driver.find_element_by_xpath(page).get_attribute('href')
                except:
                    try:
                        url=driver.find_element_by_xpath(page1).get_attribute('href')
                    except:
                        print("'네이버 뉴스'가 없습니다.")
                    
                empty=[]
                
                try:
                    req = requests.get(url, headers={'User-Agent':'Mozilla/5.0'})
                    html=req.text
                    soup = BeautifulSoup(html, 'html.parser')

                    empty.append(keywords[i])
                    title=soup.select('h3#articleTitle')[0].text                                    
                    empty.append(title)

                    text = soup.select('#articleBodyContents')[0].get_text().replace('\n', " ")
                    btext = text.replace("// flash 오류를 우회하기 위한 함수 추가 function _flash_removeCallback() {}", "")                
                    empty.append(btext)
                    print("저장 완료")
                    
                    datamart.append(empty)   #병합
                    sleep(2)
                    
                except:
                    print("url에 접근할 수 없습니다.")
                    
                sleep(1)

            except:   #네이버뉴스 포털에 뉴스가 없는 경우
                print("뉴스가 끝났습니다.")
        try:
            try:
                driver.find_element_by_xpath('/html/body/div[3]/div[1]/div[3]/div/div/button[2]').send_keys(Keys.ENTER)   #다음 페이지 이동
            except:
                driver.find_element_by_xpath('/html/body/div[3]/div[2]/div/div[1]/div[2]/div/a[2]').send_keys(Keys.ENTER)   #다음 페이지 이동
        except Exception as e:  #마지막 페이지
            print("마지막 페이지 입니다.")
            break;  
            
    driver.close()
    driver.switch_to.window(driver.window_handles[0])   #네이버 뉴스 창으로 페이지 이동
    sleep(2)

df=pd.DataFrame(datamart)   #최종 리스트 데이터프레임으로 변환
df
```
